## Ejercicios

### Ejercicio 1: Control de Flujo

Un objeto se mueve en línea recta con una velocidad constante. La posición del objeto en función del tiempo está dada por:

$$
x(t) = x_0 + v \cdot t
$$

Donde:
- $x_0$ es la posición inicial,
- $v$ es la velocidad,
- $t$ es el tiempo.

Escribe un codigo que:
1. Pida al usuario ingresar los valores de $x_0$, $v$, y un rango de tiempo $[t_{\text{inicio}}, t_{\text{fin}}]$.
2. Use un bucle `for` para calcular y mostrar la posición del objeto en intervalos de 1 segundo dentro del rango de tiempo especificado.
3. Si la posición del objeto supera los 100 metros, detén el cálculo e imprime un mensaje indicando que el objeto ha alcanzado una distancia significativa.


---

### Ejercicio 2: Funciones - Caída Libre

La altura de un objeto en caída libre en función del tiempo está dada por:

$$
y(t) = y_0 - \frac{1}{2} g t^2
$$

Donde:
- $y_0$ es la altura inicial,
- $g = 9.8 \, \text{m/s}^2$ es la aceleración debida a la gravedad,
- $t$ es el tiempo.

Escribe una función llamada `calcular_altura` que:
1. Reciba como argumentos $y_0$ y $t$.
2. Calcule y retorne la altura $y(t)$.
3. Si el tiempo $t$ es negativo, la función debe imprimir un mensaje de error y retornar `None`.

Luego, usa esta función en un condigo que:
1. Pida al usuario ingresar $y_0$ y un rango de tiempo $[t_{\text{inicio}}, t_{\text{fin}}]$.
2. Use un bucle `while` para calcular y mostrar la altura del objeto en intervalos de 0.5 segundos hasta que el objeto alcance el suelo ($y(t) \leq 0$).

### Ejercicio 3: Combinando Control de Flujo y Funciones - Lanzamiento de Proyectiles

Un proyectil es lanzado con una velocidad inicial $v_0$ y un ángulo $\theta$ respecto a la horizontal. Las ecuaciones de movimiento para el proyectil son:

$$
x(t) = v_0 \cos(\theta) \cdot t
$$
$$
y(t) = v_0 \sin(\theta) \cdot t - \frac{1}{2} g t^2
$$

Donde:
- $v_0$ es la velocidad inicial,
- $\theta$ es el ángulo de lanzamiento en radianes,
- $g = 9.8 \, \text{m/s}^2$ es la aceleración debida a la gravedad,
- $t$ es el tiempo.

Escribe un codigo que:
1. Defina una función `calcular_posicion` que reciba $v_0$, $\theta$, y $t$, y retorne las coordenadas $(x, y)$ del proyectil en ese instante.
2. Pida al usuario ingresar $v_0$, $\theta$ (en grados), y un rango de tiempo $[t_{\text{inicio}}, t_{\text{fin}}]$.
3. Use un bucle `for` para calcular y mostrar la posición del proyectil en intervalos de 0.1 segundos.
4. Detenga el cálculo si el proyectil toca el suelo ($y(t) \leq 0$).
